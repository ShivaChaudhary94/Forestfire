## Feature Selection 

```python
## Check for Multicollinearity
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr, annot=True)
```

```python
def correlation(dataset, threshold): #threshold--Means gives % 
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i , j])>threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
    
```

```python
## Threshold--- Domain expertise
corr_feature = correlation(X_train, 0.85)
```

```python
corr_feature
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_feature,axis=1,inplace=True)
X_test.drop(corr_feature,axis=1, inplace=True)
X_train.shape, X_test.shape
```

# Feature Scaling or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

# Box Plot to Understand Effect of Standard Scaler

```python
plt.subplots(figsize=(20,10))
plt.subplot(1, 2, 1) # 1 row, 2 column , 1 plot
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2) #1 row, 2 column , 2 plot
sns.boxplot(data=X_train_scaled)
plt.title('X_train after Scaling')
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score 
linreg = LinearRegression()
linreg.fit(X_train_scaled , y_train) # why we don't do tranform
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print('Mean absolute Error', mae)
print('R2 Score', score)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score1 = r2_score(y_test, y_pred)
print('Mean Absolute Error', mae)
print('R2 Score', score1)
```

# Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
score = r2_score(y_test, y_pred)
print('Mean absolute error', mae)
print('R2 Score', score)
```

# Elasticnet Regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
elastic = ElasticNet()
elastic.fit(X_train_scaled, y_train)
y_pred = elastic.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print('Mean Absolute error', mae)
print('R2 Score', score)
```
